## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.inspection import permutation_importance

RANDOM_STATE = 42
```

## EDA ,Visual Data Analysis

```python
plt.figure(figsize=(8,4))
sns.histplot(data['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Sex vs Heart Disease
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='heart_disease_present', data=data)
plt.title("Sex vs Heart Disease")
plt.xlabel("Sex (0=Female, 1=Male)")
plt.show()
```

```python
# Cholesterol vs Heart Disease
plt.figure(figsize=(8,4))
sns.boxplot(x='heart_disease_present', y='serum_cholesterol_mg_per_dl', data=data)
plt.title("Cholesterol Levels vs Heart Disease")
plt.xlabel("Heart Disease")
plt.ylabel("Cholesterol")
plt.show()
```

```python
# Chest Pain Type vs Heart Disease
plt.figure(figsize=(6,4))
sns.countplot(x='chest_pain_type', hue='heart_disease_present', data=data)
plt.title("Chest Pain Type vs Heart Disease")
plt.xlabel("Chest Pain Type")
plt.show()
```

```python
# Max Heart Rate vs Heart Disease
plt.figure(figsize=(8,4))
sns.boxplot(x='heart_disease_present', y='max_heart_rate_achieved', data=data)
plt.title("Max Heart Rate vs Heart Disease")
plt.xlabel("Heart Disease")
plt.ylabel("Max Heart Rate Achieved")
plt.show()
```

```python
# ST Depression vs Heart Disease
plt.figure(figsize=(8,4))
sns.boxplot(x='heart_disease_present', y='oldpeak_eq_st_depression', data=data)
plt.title("Oldpeak (ST Depression) vs Heart Disease")
plt.xlabel("Heart Disease")
plt.ylabel("Oldpeak ST Depression")
plt.show()
```

```python
# Major Vessels vs Heart Disease
plt.figure(figsize=(6,4))
sns.countplot(x='num_major_vessels', hue='heart_disease_present', data=data)
plt.title("Major Vessels Count vs Heart Disease")
plt.xlabel("Number of Major Vessels")
plt.show()
```

```python
# Fasting Blood Sugar vs Heart Disease
plt.figure(figsize=(6,4))
sns.countplot(x='fasting_blood_sugar_gt_120_mg_per_dl', hue='heart_disease_present', data=data)
plt.title("Fasting Blood Sugar vs Heart Disease")
plt.xlabel("FB Sugar > 120 mg/dl (1=Yes)")
plt.show()
```

```python
# Correlation Heatmap  Numeric columns only
numeric_data = data.select_dtypes(include=['int64', 'float64'])
corr = numeric_data.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
